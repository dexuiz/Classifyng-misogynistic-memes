```python
from mmf.common.registry import registry
import torch 
import numpy as np 
import os 
import pandas as pd
from tqdm.auto import tqdm
```

```python
model_cls = registry.get_model_class("mmbt")
model = model_cls.from_pretrained("mmbt.hateful_memes.images")
```

```python
test = torch.randn((1,3,224,224))
```

```python
import zipfile
with zipfile.ZipFile('/content/drive/MyDrive/computationSS/meme-test.zip', 'r') as zip_ref:
    zip_ref.extractall('/content/drive/MyDrive/computationSS/TESTING')
```

```python
TESTING_PATH  = "/content/drive/MyDrive/computationSS/TESTING/meme-test"
TESTING_CSV = os.path.join(TESTING_PATH,"Test.csv")
```

```python
test_file = pd.read_csv(TESTING_CSV,sep="\t")
```

```python
test_file.to_numpy()
```

```python
out= []

progress = tqdm(range(1000))
for row in test_file.to_numpy():
    image = row[0]
    text = row[1]
    image_path = os.path.join(TESTING_PATH,image)
    res = model.classify(image_path,text)
    output = f"{image}\t{res['label']}\n"
    print(output)
    out.append(output)
    progress.update(1)

with open("/content/drive/MyDrive/computationSS/TESTING/test_results.txt","w") as f:
    f.writelines(out)
```

```python
lines = open("/content/drive/MyDrive/computationSS/TESTING/test_results.txt","r").readlines()
```

```python
lines_ground = open("/content/drive/MyDrive/computationSS/test_labels.txt","r").readlines()
```

```python
model_outs = [int(i.split("\t")[1]) for i in lines]
```

```python
ground_outs = [int(i.split("\t")[1]) for i in lines_ground]
```

```python
from sklearn.metrics import accuracy_score,f1_score
```

```python
accuracy_score(ground_outs,model_outs)
```

```python
f1_score(ground_outs,model_outs)
```
